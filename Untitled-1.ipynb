```python
from custom_curve_signature.keygen import generate_keypair
from custom_curve_signature.hash import hash_msg
from custom_curve_signature.sign import sign
from custom_curve_signature.verify import verify
```

```python
priv, pub = generate_keypair()
priv, pub
```

```python
text1 = "Hello world ắ bờ cờ ràoofoo"
```

```python
h = hash_msg(text1.encode('utf-8'))
hex(h)
```

```python
sig = sign(h, priv)
sig

```

```python
b = hash_msg(b"Hello em Nam Nguyen ngu vcl")
hex(b)
```

```python
verify(b, sig, pub)

```

```python
verify(h, sig, pub)

```
